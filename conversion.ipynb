```python
data = pd.read_csv("delta2022.csv", skiprows=7)
```

```python
# aggregated_data = data.groupby('Destination Airport').apply(lambda x: \
#                         pd.Series({"avg_delay": x['Departure delay (Minutes)'].mean(), \
#                          "total_count": len(x)}))

def process_grobj(grobj):
    
    avg_delay = grobj['Departure delay (Minutes)'].mean()
    
    total_count = len(grobj)
    data_dict = {"avg_delay": avg_delay, "total_count": total_count}
    
    return pd.Series(data_dict)

aggregated_data = data.groupby('Destination Airport').apply(process_grobj)
```

```python
aggregated_data
```

```python
merged = pd.merge(filtered, aggregated_data, left_on='ARPT_ID', right_index=True)
```

```python
merged.sort_values('avg_delay')
```
